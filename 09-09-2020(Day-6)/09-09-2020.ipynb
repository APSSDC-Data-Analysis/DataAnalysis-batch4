## 2. Robust Scaler
- Scaling between Inter Quartile Range (25% to 75%)
    - [Xi - Q1(X)] / [Q3(X) - Q1(X)]
        - Q1 - 25% (Lower Quartile)
        - Q2 - 50% (Median)
        - Q3 - 75% (Upper Quartile)

```python
df3.describe()
```

```python
s2 = preprocessing.RobustScaler()
```

```python
s2
```

```python
s2_df = s2.fit_transform(df3)
```

```python
s2_df
```

```python
s2_df = pd.DataFrame(s2_df,columns = df3.columns)
```

```python
s2_df
```

```python
fig, (ax1,ax2) = plt.subplots(ncols = 2, figsize = (9,5))

ax1.set_title('Original Data Set')
sns.kdeplot(df3['TV'],ax = ax1)
sns.kdeplot(df3['radio'],ax = ax1)
sns.kdeplot(df3['newspaper'],ax = ax1)
sns.kdeplot(df3['sales'],ax = ax1)

ax2.set_title('After Robust Scaling')
sns.kdeplot(s2_df['TV'],ax = ax2)
sns.kdeplot(s2_df['radio'],ax = ax2)
sns.kdeplot(s2_df['newspaper'],ax = ax2)
sns.kdeplot(s2_df['sales'],ax = ax2)
```

### 3. MinMax Scaler
- Scaling between min and max value of our data set
    - [Xi - min(X)] / [max(X) - min(X)]

```python
s3 = preprocessing.MinMaxScaler()
```

```python
s3
```

```python
s3_df = s3.fit_transform(df3)
```
