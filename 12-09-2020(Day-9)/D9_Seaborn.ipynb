

# Data Visualization using Seaborn

    * Seaborn is a Python data visualization library based on matplotlib
    * It provides a high-level interface for drawing attractive and informative statistical graphics
    * Seaborn is a library for making statistical graphics in Python
    * Applications - used in visualising data in Machine learning, data Science 
                   - statistical aggregation to produce informative plots
1. Color Palette
    * This function provides an interface to many (though not all) of the possible ways you can generate colors in seaborn

2. Plotting with categorical data

    **Categorical scatterplots:**

        * stripplot() (with kind="strip"; the default)

        * swarmplot() (with kind="swarm")

    **Categorical distribution plots:**

        * boxplot() (with kind="box")

        * violinplot() (with kind="violin")
3. Joint plot
4. Pairplot
5. Heatmap


```python
import seaborn as sns
```

```python
sns.palplot(sns.color_palette())
# palette -- possibile ways to generate colors in seaborn 
```

```python
# sequencial colors 
sns.palplot(sns.color_palette("Blues"))
```

```python
sns.palplot(sns.color_palette("Blues_d"))
```

```python
# dark colors
sns.palplot(sns.dark_palette("purple"))
```

```python
sns.palplot(sns.light_palette("navy"))
```

```python
sns.palplot(sns.light_palette("navy",reverse=True))
```

```python
# diverging colors 
sns.palplot(sns.color_palette("Purples",100))
```

```python
sns.palplot(sns.hls_palette(n_colors = 15,l = 0.6,s = 0.4))
# hls_palette - color space
# l - lightness/ brighness
# hue - color shape
# s - saturation - intensity
```

# Categorical scatterplots 
* work on categorical data (species)
* default it returns scatter

```python
sns.catplot(x = "species", y = "sepal_width" , data = iris)
```

# Swarm plot
* prevents data overlapping 

```python
sns.catplot(x = "species", y = "sepal_width" , data = iris, kind = "swarm")
```

```python
sns.catplot(y = "species", x = "sepal_width" , data = iris, kind = "swarm")
```

# Box Plot
* Distribution  along with extreme values 
* min value, 25% and median, 75% and max values, outliers

```python
sns.catplot(y = "species", x = "sepal_length" , data = iris,kind = "box")
```

```python
sns.catplot(x = "species", y = "sepal_length" , data = iris,kind = "box",palette = ["m","c","g"])
```

```python
sns.catplot(data = iris, kind = "box") # it takes only numerical data 
```

```python
sns.catplot(data = iris, kind = "box" , orient = "h") # v & h
```

# Violinplots
* Distribution plot
* combinations of both boxplot, kde plot 

```python
sns.catplot(x = "species", y = "sepal_length" , data = iris,kind = "violin")
```

```python
sns.violinplot(x = iris.species, y = iris.sepal_width)
```

```python
help(sns.violinplot)
```

# Joint Plot
* Combination of two plots
* default is scatter n histogram
* used to find relation between variables 

```python
sns.jointplot(x = "petal_width", y = "sepal_width",data = iris)
```

```python
sns.jointplot(x = "petal_width", y = "sepal_width",data = iris,kind = "reg") # linear regression libe
```

```python
sns.jointplot(x = "petal_width", y = "sepal_width",data = iris,kind = "kde") # linear regression libe
```

```python
help(sns.jointplot)
```

```python
sns.jointplot(x = "petal_width", y = "sepal_width",data = iris,kind = "hex")
```

# Pairplot
* used for multiple plots at a time  
* it gives pairwise relationships 
* default scatter plot
* matrix plot

```python
sns.pairplot(iris, hue = "species")
# hue - color based on species 
```

```python
sns.pairplot(iris, hue = "species",kind= "scatter")
```

```python
sns.pairplot(iris, hue = "species",kind= "scatter",diag_kind = "kde",markers =["^","d","o"])
# hue - categorical data from specified column
```

```python
help(sns.pairplot)
```

# Heat map
* some sort matrix plot
* used to find correlation between features 
* correlation - impact of one on another / connection between variables 

```python
tip = sns.load_dataset("tips")
tip
```

```python
tips_cor = tip.corr()
tips_cor
```

```python
sns.heatmap(tips_cor)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize = (10,8))
sns.heatmap(tips_cor,annot = True)
```

```python
sns.heatmap(tips_cor,annot = True,linecolor = "black", linewidth = 2,cmap = "plasma")
```
